```python
plt.figure(1,figsize=(15, 6))
n = 0
for x in ["Age",	"Annual Income (k$)" ,	"Spending Score (1-100)"]:
    n += 1
    plt.subplot( 1,3,n)
    plt.subplots_adjust(hspace=0.5, wspace=0.5)
    sns.distplot(df[x], bins=20, kde=True)  # Use histplot with kde=True for smooth density curve
    plt.title('Distplot of {}'.format(x))
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(y="Gender",data=df)
plt.show()
```

```python
plt.figure(1,figsize=(15,7))
n=0
for x1 in ["Age","Annual Income (k$)","Spending Score (1-100)"]: 
    # plt.subplot (1,3,n)
    n=n + 1
    plt.subplot (1,3,n)
    sns.set(style="whitegrid")
    plt.subplots_adjust(hspace=0.5,wspace=0.5)
    sns.violinplot(x=x1,y="Gender",data=df)
    plt.ylabel("Gender" if n==1 else " ")
plt.show()
    
    
```

```python
agex=["age_18_25","age_26_35","age_36_45","age_46_55","age_55"]
agey=[len(age_18_25.values),len(age_26_35.values),len(age_36_45.values),len(age_46_55.values),len(age_55.values)]
plt.figure(figsize=(7,10))
sns.barplot(x=agex,y=agey)
plt.title("number of customer and ages")
plt.xlabel("age")
plt.ylabel("values")
plt.show()
```
